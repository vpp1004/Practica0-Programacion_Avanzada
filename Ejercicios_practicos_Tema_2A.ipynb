### Ejercicio 1. Creación de un Dot Plot Simple.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
def dot_plot(seq1, seq2):
    long1 = len(seq1)
    long2 = len(seq2)

    matriz = np.zeros((long1, long2))

    for i in range(long2):
        for j in range(long1):
            if seq2[i] == seq1[j]:
                matriz[i][j] = 1 # Coincidencia

    plt.figure(figsize=(6,6))
    plt.imshow(matriz, cmap="binary")

    plt.xticks(ticks=np.arange(len(seq2)), labels=list(seq2))
    plt.yticks(ticks=np.arange(len(seq1)), labels=list(seq1))

    plt.xlabel("Secuencia 2")
    plt.ylabel("Secuencia 1")
    plt.title("Dot Plot")
    plt.show()
```

```python
dot_plot("GATTACA", "GCATGCU")
```
